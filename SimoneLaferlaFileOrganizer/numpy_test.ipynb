```python
import os,shutil
from PIL import Image
import numpy as np
from tabulate import tabulate
import csv

directory_principale = "C:/Users/SLK/Desktop/FileOrganizer/files"
directory_imm = "C:/Users/SLK/Desktop/FileOrganizer/files/imm"


def WriteToCsv(msg):
    csvPath = os.path.join(directory_principale, "recap.csv")
    with open(csvPath, mode = "a", newline='') as file:  
        writer = csv.writer(file)
        writer.writerow([msg])
        print(msg)

os.getcwd()


contenuti = os.listdir(directory_principale)
print("contenuti del file sono:",contenuti)
# directory per creare le cartelle
cartella_audio = os.path.join(directory_principale, "Aud")
cartella_documenti = os.path.join(directory_principale, "Docum")
cartella_immagini = os.path.join(directory_principale, "Imm")

os.makedirs(cartella_audio, exist_ok=True)
os.makedirs(cartella_documenti, exist_ok=True)
os.makedirs(cartella_immagini, exist_ok=True)
#All'inizio avevo usato solo makedirs per creare le cartelle e inserire i file al loro interno, e funzionava.
#Successivamente, però, ha iniziato a dare errore, quindi ho dovuto chiedere a ChatGPT il motivo. Mi ha suggerito di creare il percorso in questo modo: cartella_audio = os.path.join(directory_principale, "Aud.").
#Perché prima funzionava e dopo no?
print("Cartelle create.")

#  ciclo for per cercare ogni estenzione dei file per inserirli nelle cartelle giuste 
for file in os.listdir(directory_principale):
    percorso_file = os.path.join(directory_principale, file)
    sizeFile = str(os.stat(percorso_file).st_size)
    
    if file.endswith((".png", ".jpg", ".jpeg")):
        shutil.move(percorso_file,cartella_immagini)
        msg = file + " type: image size: " + sizeFile
        WriteToCsv(msg)# richiama la funzione 
    elif file.endswith((".odt", ".txt")):
        shutil.move(percorso_file, cartella_documenti)
        msg = file + " type: text size: " + sizeFile
        WriteToCsv(msg)
    elif file.endswith(".mp3"):
        shutil.move(percorso_file,cartella_audio)
        msg = file + " type: audio size: " + sizeFile
        WriteToCsv(msg)

#creazione di una lista per raccogliere i dati 
dati = []

# controllare quali sono i fali da mostrare sulla tabella 
for file in os.listdir(directory_imm):
    if file.lower().endswith((".png", ".jpg", ".jpeg")):
        percorso = os.path.join(directory_imm, file)
        immagine = Image.open(percorso)
        
        
        arr = np.array(immagine)
        
# da qui in poi mi sono fatto aiutare, non sarei mai riusito da solo, mi sarebbe piaciuto avere piu basi solide per fare questo test,
# e non perdere tempo a cercare online per poi non riuscire a collegare i pezzi e non capire come fare, sicurmente e essenziale saper cercare online, ma prima bisognerebbe creae una base solita per capire che cosa si fa.
        if len(arr.shape) == 2:
            
            grayscale_media = np.mean(arr)
            dati.append([file, immagine.height, immagine.width, grayscale_media, None, None, None, None])
        elif len(arr.shape) == 3:
            if arr.shape[2] == 3:
               
                media = np.mean(arr, axis=(0, 1))
                r_media, g_media, b_media = media
                dati.append([file, immagine.height, immagine.width, None, r_media, g_media, b_media, None])
            elif arr.shape[2] == 4:
                
                media = np.mean(arr, axis=(0, 1))
                r_media, g_media, b_media, a_media = media
                dati.append([file, immagine.height, immagine.width, None, r_media, g_media, b_media, a_media])
        
        immagine.close()
#creazione della tabella 
intestazioni = ["Nome File", "Altezza", "Larghezza", "Grayscale", "R (Red)", "G (Green)", "B (Blue)", "A (Alpha)"]
tabella = tabulate(dati, headers=intestazioni, tablefmt="grid")
print(tabella)
    
```
