```python
#alimenti è una lista che contiene tutti gli alimenti che ci verranno scritti dal cliente 
alimenti = []

# Il codice inizia con una macro funzione chiamata "market", dove si chiederà al cliente di inserire i dati richiesti tramite il ciclo while.
def market():
    print("Benvenuto nel market56")        

    while True: # Ciclo principale per il menu interattivo.

        print("Premi 1 per inserire un prodotto")
        print("Premi 2 per visualizzare tutti i prodotti")
        print("Premi 3 per ricercare un prodotto")
        print("Premi 4 per visualizzare statistiche sugli alimenti")
        print("Premi 0 per uscire")

        cmd = input()

        if cmd == "0":
            print("Grazie per aver utilizzato market56")
            break
        elif cmd == "1":
            inserimento()
        elif cmd == "2":
            visualizzazione()
        elif cmd == "3":
            ricerca()
        elif cmd == "4":
            statistica()

# Questa funzione chiede al cliente di inserire determinati dettagli di un prodotto che ci aiuteranno in futuro a soddisfare le richieste del cliente.
def inserimento():
    cmd = "1"
    while cmd == "1":
        print("ID prodotto da aggiungere")
        id = input()
        print("Nome prodotto da aggiungere: ")
        nome = input()
        print("Quantità prodotto da aggiungere")
        qt = input()
        print("Categoria prodotto da aggiungere: ")
        categoria = input()
        print("Ingredienti prodotto da aggiungere (separati da virgola): ")
        ingredienti = input().replace(" ", "").split(",")
        print("Tempo di preparazione del prodotto da aggiungere (in minuti): ")
        tempo = input()
# Questa condizione verifica se un alimento con lo stesso nome è già presente.
        trovato = False
 
        for alimento in alimenti:
            if alimento.nome == nome:
                trovato = True
                print("L'elemento esiste già")
        
        if not trovato:
            alimenti.append(Food(id, nome, qt, categoria, ingredienti, tempo))
        
        print("Premi 1 per inserire un nuovo prodotto")
        print("Premi 0 per uscire")
        cmd = input()
     

# La funzione visualizzazione mostrerà l'elenco degli alimenti, con diverse tipologie di filtro (id, nome, quantità ecc.) e, in caso non ci siano prodotti, stamperà a video ("Non ci sono prodotti").
def visualizzazione():
    if len(alimenti) > 0:
        print("\nElenco prodotti inseriti:")
        for prodotto in alimenti:
            print("ID:", prodotto.id, 
                    "- Nome:", prodotto.nome, 
                    "- Quantità:", prodotto.qt,
                    "- Categoria:", prodotto.category,
                    "- Tempo:", prodotto.tempo, 
                    "- Ingredienti:", ", ".join(prodotto.ingredienti))
    else:
        print("Non ci sono prodotti")


# Funzione di ricerca per gestire diverse modalità di ricerca di alimenti.
# L'utente può scegliere tra 4 tipi di ricerca:
# - Per nome (opzione 1)
# - Per categoria (opzione 2)
# - Per ID (opzione 3)
# - Ricerca avanzata con filtri (opzione 4)
# Dopo ogni ricerca, l'utente può decidere se continuare o tornare al menu principale.
def ricerca():
    cmd = "1"
    while cmd == "1":
        print("Premi 1 per ricerca per nome")
        print("Premi 2 per ricerca per categoria")
        print("Premi 3 per ricerca per ID")
        print("Premi 4 per ricerca avanzata")
        scmd = input()

        if scmd != "4":
            print("Inserire valore da ricercare")
            value = input()

            for alimento in alimenti:
                if scmd == "1" and alimento.nome == value:
                    print("ID:", alimento.id, 
                        "- Nome:", alimento.nome, 
                        "- Quantità:", alimento.qt,
                        "- Categoria:", alimento.category,
                        "- Tempo:", alimento.tempo, 
                        "- Ingredienti:", ", ".join(alimento.ingredienti))
                elif scmd == "2" and alimento.category == value:
                    print("ID:", alimento.id, 
                        "- Nome:", alimento.nome, 
                        "- Quantità:", alimento.qt,
                        "- Categoria:", alimento.category,
                        "- Tempo:", alimento.tempo, 
                        "- Ingredienti:", ", ".join(alimento.ingredienti))
                elif scmd == "3" and str(alimento.id) == value:
                    print("ID:", alimento.id, 
                        "- Nome:", alimento.nome, 
                        "- Quantità:", alimento.qt,
                        "- Categoria:", alimento.category,
                        "- Tempo:", alimento.tempo, 
                        "- Ingredienti:", ", ".join(alimento.ingredienti))
                else:
                    print("Nessun alimento trovato.")

         # Ricerca avanzata con filtri (tempo massimo, ingredienti e categoria)
        else:
            print("Inserire il tempo massimo di preparazione: ")
            tempo_massimo = int(input())
            print("Inserire gli ingredienti richiesti (separati da virgola): ")
            ingredienti_richiesti = input().replace(" ", "").split(",")
            print("Inserire la categoria desiderata (lasciare vuoto per non filtrare): ")
            categoria = input()   

# Applica i filtri con la funzione `filtra_alimenti`.
            risultati = filtra_alimenti(alimenti, tempo_massimo, ingredienti_richiesti, categoria)

            if risultati:
                print("Alimenti trovati:")
                for prodotto in risultati:
                    print(f"Nome: {prodotto.nome}, Tempo: {prodotto.tempo}, Ingredienti: {', '.join(prodotto.ingredienti)}")
            else:
                print("Nessun alimento trovato.")

# Mostra il menu per continuare o tornare al menu principale.
        print("Premi 1 per cercare un altro alimento")
        print("Premi 0 per tornare al menu principale")

        cmd = input()


# `filtra_alimenti` mostra tramite delle condizioni impostate dal cliente una lista filtrata.
# La funzione restituisce una lista di alimenti che soddisfano tutti i criteri forniti.
def filtra_alimenti(alimenti, tempo_massimo=None, ingredienti_richiesti=None, categoria=None):
        risultati = []
        for alimento in alimenti: # Se l'alimento soddisfa i criteri, verrà aggiunto alla lista dei risultati.
            if tempo_massimo is not None and alimento.tempo > tempo_massimo: # Esclude alimenti con tempo maggiore di quello specificato.
                continue
            if ingredienti_richiesti is not None:
                if not any(ingrediente in alimento.ingredienti for ingrediente in ingredienti_richiesti):
                    continue
            if categoria and alimento.category != categoria: # Esclude alimenti non appartenenti alla categoria specificata.
                continue
            risultati.append(alimento) # Il risultato viene aggiunto alla lista "risultati".
        return risultati


# Crea statistiche sugli alimenti in modo che il cliente possa avere maggiori informazioni sulla propria ricerca.
def statistica():
    cmd = "1"
    while cmd == "1":  
    # Menu di scelta per l'utente.
        print("Premi 1 per il conteggio degli alimenti")
        print("Premi 2 per il conteggio degli alimenti per categoria")
        print("Premi 0 per uscire")

        try:
    # Richiesta della scelta all'utente e conversione in intero.
            scmd = int(input("Inserisci la tua scelta: "))
        except ValueError:
            print("Errore: Inserisci un numero valido.")
            continue
    # Crea un conteggio degli alimenti nella lista.
        if scmd == 1:
            print(f"Numero totale di alimenti: {len(alimenti)}")
        elif scmd == 2:
            categories = {}

            for alimento in alimenti:
                if alimento.category in categories:
                    categories[alimento.category] += 1
                else:
                    categories[alimento.category] = 1
        # Stampa i valori trovati, ossia la categoria e il numero di alimenti appartenenti.
            for key, value in categories.items():
                print("La categoria: " + key + " ha " + str(value) + " elementi")


class Food:
# Questi sono i dati fondamentali che il cliente deve fornire per aggiungere altri alimenti.
      def __init__(self, id, nome, qt, category, ingredienti, tempo):
          self.id = id
          self.nome = nome
          self.qt = qt
          self.category = category
          self.ingredienti = ingredienti  # Deve essere una lista.
          self.tempo = int(tempo) 


# Questi sono elementi aggiuntivi inseriti nella lista in modo che il cliente abbia già una base per ricercare e trovare alcune ricette ed alimenti.
alimenti.append(Food(1, "pasta", 15, "pasta", "farina, acqua".replace(" ", "").split(","), 5))
alimenti.append(Food(2, "ragù", 7, "preparati", "carne, pomodoro,olio,sale,pepe".replace(" ", "").split(","), 5))
alimenti.append(Food(3, "pesto", 5, "preparati", "basilico,olio,aglio,sale,pepe,pinoli".replace(" ", "").split(","), 5))
alimenti.append(Food(4, "pizza", 20, "forno", "farina, acqua,lievito, pomodoro,basilico,".replace(" ", "").split(","), 5))
alimenti.append(Food(5, "uova", 20, "frigo", "uova,".replace(" ", "").split(","), 5))

market()
```
